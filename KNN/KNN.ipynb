# K-Nearest Neighbors (KNN) Classification from Scratch

## Overview
Implementation of the K-Nearest Neighbors (KNN) algorithm from scratch using NumPy and Matplotlib, including:
- Synthetic dataset generation
- Noise injection
- Euclidean distance computation
- KNN prediction
- Accuracy evaluation using Leave-One-Out Cross Validation (LOOCV)
- Decision boundary visualization
- Analysis of overfitting and underfitting for different K values

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generating data
def gen_data():
    np.random.seed(42)
    n=100
    class0=np.random.randn(n,2)*1.5 + np.array([2,2])
    class1=np.random.randn(n,2)*1.5 + np.array([4,4])
    X=np.vstack((class0,class1)) #to merge using rows
    Y=np.array([0]*n+[1]*n)

    noise_ratio=0.1
    num_noisy=int(noise_ratio*len(Y))
    noise_idx=np.random.choice(len(Y),num_noisy,replace=False)
    Y[noise_idx]=1-Y[noise_idx]
    return X,Y
```

```python
#distance function
def euclidean_distance(a,b):
    return np.sqrt(np.sum((a-b)**2))
```

```python
#KNN prediction
def knn_pred(X_train,Y_train,X_test,k=3):
    dist=[]
    for i in range(len(X_train)):
        d=euclidean_distance(X_test,X_train[i])
        dist.append((d,Y_train[i]))

    dist.sort(key=lambda x:x[0])
    k_nearest=dist[:k]
    labels=[label for _,label in k_nearest]
    preds=max(set(labels),key=labels.count)
    return preds
```

```python
def plot_data(X,y):
    plt.scatter(X[:,0],X[:,1],c=y)
    plt.title("Dataset")
    plt.show();

def plot_decision_boundary(X,y,k=3):
    x_min,x_max = X[:,0].min()-1, X[:,0].max()+1
    y_min,y_max = X[:,1].min()-1, X[:,1].max()+1

    xx,yy=np.meshgrid(
        np.linspace(x_min, x_max, 100),
        np.linspace(y_min, y_max, 100)
    )
    z=np.zeros(xx.shape)
    for i in range(xx.shape[0]):
        for j in range (xx.shape[1]):
            point=np.array([xx[i,j],yy[i,j]])
            z[i,j]=knn_pred(X,y,point,k)
            
    plt.contourf(xx, yy, z, alpha=0.3)
    plt.scatter(X[:,0], X[:,1], c=y)
    plt.title(f"KNN Decision Boundary (K={k})")
    plt.show()
```

# Observations:-
- with K = 1 we get extreme over fitting with 67 accuracy
- with k = 15 we get an almost straight line and a very high accuracy
- with k = 35, there is more straight line but the accuracy decreased (beginning of underfitting)
- with k = 50, increases the accuracy again and line loses its curves more

Overall, smaller K values tend to overfit the data, while larger K values produce smoother and more generalized decision boundaries.
